```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

daily_demand = df.groupby(df['InvoiceDate'].dt.date)['Quantity'].sum()

weekly_demand = df.groupby(
    pd.Grouper(key='InvoiceDate', freq='W')
)['Quantity'].sum()

monthly_demand = df.groupby(
    pd.Grouper(key='InvoiceDate', freq='ME')
)['Quantity'].sum()

display(daily_demand.head())
display(weekly_demand.head())
display(monthly_demand)
```
