Next I'll create RFM features. Recency, Frequency, and Monetary. They represent, respectively, days since last purchase, number of invoices and total revenue. RFM is a relevant step to K-Means because it converts raw transaction data into numerical customer behaviour features. RFM creates one profile per customer using Recency, Frequency and Monetary to measure how much they spent. These variables are important to K-Means because the algorithm groups customers based on numerical similarity, what allows the identification of segments such as high value loyal customers, recent active customers, occasional buyers, etc. 

First I'll create a reference date to use one day after the last transactions. Them I'll group all transactions belonging to the same customer. Recency will find each customer's most recent purchase and calculates how many days have passed since then. Frequency will count how many unique invoices each customer has. Monetary will calculate the total amount spent by each customer. 

```python
backup_date = customer_df["InvoiceDate"].max() + pd.Timedelta(days=1)
```

```python
rfm_df = customer_df.groupby("Customer ID").agg(
    Recency=("InvoiceDate", lambda x: (backup_date - x.max()).days),
    Frequency=("Invoice", "nunique"),
    Monetary=("Revenue", "sum")
).reset_index()

print("RFM dataset shape: ", rfm_df.shape)
rfm_df.head()
```

The original customer transaction dataset contained 400,916 rows because each row represented an individual product purchase within an invoice. For customer segmentation, the data needed to be aggregated to make each row represent one customer instead of one transaction. After grouping by Customer ID and calculating Recency, Frequency, and Monetary values, the final RFM dataset contains 4,312 customer profiles and 4 columns. This structure is right for K-Means because each customer is now represented by numerical behavioural features.

Next I have to check the RFM dataset to understand it. 

```python
rfm_df.describe()
```

```python
rfm_df.isnull().sum()
```

```python
rfm_df.sort_values(by="Monetary", ascending=False).head(10)
```

It's verifiable that the RFM dataset is highly skewed, it supports the idea of VIP customers who spend much more than regular customers. K-Means is distance based to if I feed the raw values directly into K-Means, the Monetary feature will dominate the clustering because its values are much bigger than Recency and Frequency. So to fix it, I'll apply log transformation to reduce skewness and scale the features so they are comparable. 

I've decided to use log1p() as the best option because the RFM variables are very right skewed, meaning that most customers spend relatively little, but a small number of customer spend extremely large amounts. Since K-Means is distance based, these extreme values would dominate the clustering and make the model focus too heavily on a few VIP customers. Log1p then reduces this problem by compressing large values more than small ones while still keeping the rank of customers.

```python
rfm_model = rfm_df[["Recency", "Frequency", "Monetary"]].copy()

rfm_log = rfm_model.copy()

rfm_log["Recency"] = np.log1p(rfm_log["Recency"])
rfm_log["Frequency"] = np.log1p(rfm_log["Frequency"])
rfm_log["Monetary"] = np.log1p(rfm_log["Monetary"])

rfm_log.describe()
```

I had to decide among a few scaling options and I've chosen Standard Scaler. My choice was based on my final goal, that is to apply K-Means. K-Means is a distance based algorithm and requires the input variables to be measured on comparable scales. After applying the log1p() to reduce the strong skew in the RFM variables I'll use Standard Scaler to centre all features around zero with a standard deviation of one. This will prevent MOnetary or Frequency from dominating the clustering just because they have a larger numerical range. I've chosen this scaling method over MinMax, RobustScaler, Normalizer etc because the StandardScaler provides the best balance for this dataset as the main outlier impacting the dataset was already reduced through log transformation. 

The scaled RFM DataFrame will convert the transformed Recency, Frequency, and Monetary values into a standardised format suitable for K-Means. The customer ID is going to be kept as the index for identification and only the three behavioural features are used for clustering. Negative values will be present now, meaning some customers are below the average for that feature after standardisation. 

```python
scaler = StandardScaler()
```

```python
rfm_scaled = scaler.fit_transform(rfm_log)

rfm_scaled_df = pd.DataFrame(
    rfm_scaled,
    columns=rfm_log.columns,
    index=rfm_df["Customer ID"]
)

print("Scaled RFM shape: ", rfm_scaled_df.shape)
rfm_scaled_df.head()
```

**MODELLING PHASE**

*K-Means Model*

Next I'll perform the K-Means Elbow Method to help me decide the best number os clusters. The Elbow method tests several values of k and calculates the inertia for each one. Inertia means how tighlty customers are grouped around their cluster centre. Lower inertia means customers are closer to their assigned cluster centre, so the clusters are more compact. But, inertia decreases as more clusters are added. The goal is not just choose the lowest inertia, but to find a point where the improvement starts slowing down, finding the elbow. 

First, I'll create an empty list to store inertia values. Then I'll test different numbers of clusters and plot elbow curve. 

```python
inertia_values = []

k_range = range(2,11)

for k in k_range:
    kmeans = KMeans(
        n_clusters = k,
        random_state = 42,
        n_init = 10
    )

    kmeans.fit(rfm_scaled_df)

    inertia_values.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,6))
plt.plot(k_range, inertia_values, marker="o")
plt.xlabel("Number of Clusters k")
plt.ylabel("Inertia")
plt.title("Elbow Method for K-Means")
plt.show()
```

The Elbow method showed that the inertia decreased as the number of clusters increased, what was expected because more clusters reduce withing cluster distance. However, the curve started to flatten after 5 or 6 clusters, it suggests that additional clusters beyond this point provided smaller improvements. k=5 and k=6 were identified as candidate values for further assessment using silhouette score and business interpretability. 

Next I'll perform the Silhouette score to measure how well the K-Means clusters are separated and how consistent the customers are within their assigned clusters.  First I need to create an empty list to store the scores, and test cluster numbers from k=2 to k=10. For each value of k, K-Means assigns every customer to a cluster using fit_predict(). Then silhouette score will compare how close each customer is to its own cluster versus other clusters. Then the score will be stored and plotted so I can verify which number of clusters gives the best separation. A higher silhouette score usually means better defined clusters, but I'll still need to balance the score with business interpretability. 

```python
silhouette_scores = []

k_range = range(2,11)

for k in k_range:
    kmeans = KMeans(
        n_clusters = k,
        random_state = 42,
        n_init = 10
    )

    cluster_labels = kmeans.fit_predict(rfm_scaled_df)

    score = silhouette_score(rfm_scaled_df, cluster_labels)

    silhouette_scores.append(score)
```

```python
plt.figure(figsize=(10,6))
plt.plot(k_range, silhouette_scores, marker="o")
plt.xlabel("Number of Clusters k")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score for K-Means")
plt.show()
```

```python
print(silhouette_scores)
```

These scores compare how well separated the clusters are for each value of k. Scores close 1 mean clusters are very well separated. Scores close to 0 mean clusters overlap. Scores bellow 0 mean some points may be assigned to the wrong cluster. 
From the results, the best technical score is k=2, silhouette=0.422. It means 2 clusters are the cleanest separation, mathematically speaking. However, for my project objective having only 2 clusters is very simplistic, it would only separate the clusters between high-value/low-value. But it's possible to extract more business insights. I'd like to have at least 3-4 clusters to have a better visualization and business interpretation. So next I'll need to compare the cluster profiles. Next I'll run K-means with k=3, k=4, and k=5 and then compare the cluster profiles using original RFM values. That should show which cluster solution makes the most business sense. 

```python
candidate_k_values = [3,4,5]

kmeans_comparison = []

for k in candidate_k_values: 
    kmeans = KMeans(
        n_clusters = k,
        random_state = 42,
        n_init = 10
    )

    cluster_labels = kmeans.fit_predict(rfm_scaled_df)
    
    inertia = kmeans.inertia_

    silhouette = silhouette_score(rfm_scaled_df, cluster_labels)

    kmeans_comparison.append({
        "k" : k, 
        "inertia" : inertia,
        "silhouette_score" : silhouette
    })

kmeans_comparison_df = pd.DataFrame(kmeans_comparison)

kmeans_comparison_df
```

Clusters k=3 and k=4 are very close in cluster separation. k=5 has a lower inertia but also a weaker silhouette score, that means the model becomes more detailed but the clusters may be less clearly separated. Next I'll inspect the actual cluster profiles, so I can extract and produce business meaningful groups. 

Next I'll compare K-Means models using those 3 possibiblities. For each value of k, I'll create a K-Means model, fit it to the scaled RFM dataset and assign each customer to a cluster. Then the code will copy the original rfm_df dataset so the clusters can be interpreted using real business values instead of scaled numbers. After that, labels will be added and group customers by clusters and calculate the number of customers, average recency, average frequency, and average monetary value for each group. This way I'll be able to compare the technical performance of each model and if the clusters make sense from a business point of view.  

```python
candidate_k_values = [3, 4, 5]

for k in candidate_k_values:
    
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    rfm_profile = rfm_df.copy()
    
    rfm_profile["Cluster"] = kmeans.fit_predict(rfm_scaled_df)

    cluster_summary = rfm_profile.groupby("Cluster").agg(
        Customer_Count=("Customer ID", "count"),
        Avg_Recency=("Recency", "mean"),
        Avg_Frequency=("Frequency", "mean"),
        Avg_Monetary=("Monetary", "mean")
    ).round(2)
    
    print("K-Means Cluster Profile for k =", k)
    print(cluster_summary)
```

Based on the scores and averages. k=4 provides almost the same silhouette score as k=3, better business detail than k=3, and a clearer separation than 5=5. It creates four interpretable customer groups. Lower recency means customer bought more recently, higher frequency means the customer buys more often, and higher monetary means customer spends more. 
So I'll go with k=4 as my final K-Means model solution because it provided the strongest balance between technical performance and business interpretability. 
The interpretation of k=4 would be something like: cluster 0= Recent low frequency customers, cluster 1= High-value loyal customers, cluster 2= Inactive low-value customers, and cluster 3= Regular medium value customers.

Next I'll build the final K-Means model using k=4 and attach the cluster labels to the RFM dataset.

```python
kmeans_finalModel = KMeans(
    n_clusters=4,
    random_state=42,
    n_init=10
)

rfm_df["Cluster"] = kmeans_finalModel.fit_predict(rfm_scaled_df)

rfm_df.head()
```

Displaying cluster profiles:

```python
final_cluster_profile = rfm_df.groupby("Cluster").agg(
    Customer_Count=("Customer ID", "count"),
    Avg_Recency=("Recency", "mean"),
    Avg_Frequency=("Frequency", "mean"),
    Avg_Monetary=("Monetary", "mean")
).round(2)

final_cluster_profile
```

Now I'll define the names for all 4 segments. 

```python
segment_names = {
    0: "Recent low-frequency customers",
    1: "High-value loyal customers",
    2: "Inactive low-value customers",
    3: "Regular medium-value customers"
}

rfm_df["Segment"] = rfm_df["Cluster"].map(segment_names)

rfm_df.head(7)
```

```python
cluster_profile = rfm_df.groupby("Segment").agg(
    Avg_Recency=("Recency", "mean"),
    Avg_Frequency=("Frequency", "mean"),
    Avg_Monetary=("Monetary", "mean")
).round(2)

cluster_profile
```

After that it's possible to visualize how the dataset was re-organized and now presents the segment column separating the customers. Next I'll provide 2 visualizations so the business people can understand better what's going on.

The chosen k=4 model provided the best balance between technical performance and business interpretability, as all four groups were clearly different behavioural segments: high value loyal customers, regular medium value customers, recent low frequency customers and inactive low value customers. *Apriori Model*

Time to further develop the Apriori Model. First I'll start by generating frequent itemsets. Using Apriori on my final basket matrix (previously developed). The basket_matrix_encoded dataset contains invoices as rows and products as columns, with 1=product bought and 0=not bought. I'll set the minimum support to 0.03 so only products combinations that appear in at least 3% of invoces are used. 

```python
frequent_itemsets = apriori(
    basket_matrix_encoded,
    min_support = 0.03,
    use_colnames = True
).sort_values(
    "support",
    ascending=False
).reset_index(drop=True)
```

Next I'll create a new column called length to count how many products are inside each itemset and display the result. 

```python
frequent_itemsets["length"] = frequent_itemsets["itemsets"].apply(lambda x: len(x))

print("Number of frequent itemsets: ", len(frequent_itemsets))

frequent_itemsets.head(100)
```

Now I need to check for frequent product pairs. It's an important step because association rules need combinations of at least two products.

```python
frequent_itemsets[
    (frequent_itemsets["length"] == 2) & (frequent_itemsets["support"] >= 0.03)
]
```

Next I need to turn the frequent itemsets into association rules. Sorting by lift puts the strongest product relationships at the top.

```python
rules = association_rules(
    frequent_itemsets,
    metric="lift",
    min_threshold=1
).sort_values(
    "lift",
    ascending=False
).reset_index(drop=True)

print("Number of association rules: ", len(rules))

rules.head(10)
```

I'll make a clean Apriori rules table just for the sake of visualization in the report. 

```python
apriori_results = rules[
    [
        "antecedents",
        "consequents",
        "support",
        "confidence",
        "lift"
    ]
].copy()

apriori_results.head(10)
```

121 frequent itemsets and 18 association rules were found. The min_support threshold worked well by avoiding too many weak rules, but still produced useful product relationships. The strongest rule was between 'WOODEN PICTURE FRAME WHITE FINISH' and 'WOODEN FRAME ANTIQUE WHITE', with a lift of 10.15 and confidence of 64.7%, what indicates a reasonably strong product association.Other rules also demonstrated meaningful relationships between similar or complementary products, like ceramic trinket boxes, wicker hearts, cake cases, and decorative word blocks. These results support the upselling and recommendation objective of the project by identifying products that could be promoted together, bundled or placed near each other to encourage additional purchases. The Apriori model was used to identify frequent itemsets and association rules are interpreted mostly through support, confidence and lift. Support means how often a product or product combination appears in the whole dataset. Confidence means how often the second product is bought when the first product is bought as well. Lift measures if the relationship is stronger than random chance. Lift is usually the most important metric in Apriori because it shows the strength of the association. With that, the Modelling phase is concluded.
